```python
import pandas as pd
import numpy as np

df2 = pd.read_csv('f1data.csv')
```

```python
df2

#take away unneeded data
df2.columns.str.match("Unnamed")
df2.drop('constructorId',inplace=True,axis=1)
df=df2.loc[:,~df2.columns.str.match("Unnamed")]
#set year as int
#col = ['year']
#df[col] = df[col].astype(int)
df
```

```python
def convert(current, msf):
    try:
        minutes, seconds = msf.split(':')
        seconds, milliseconds = seconds.split('.')
        minutes, seconds, milliseconds = map(int, (minutes, seconds, milliseconds))
        return (minutes * 60 + seconds) * 1000 + milliseconds
    except:
        print(current, msf)
```

```python
#pivot for ease of use and reading; by year and circuitName
pivot_table_df = df.pivot(
    index='year',
    columns='circuitName',
    values='bestLapTime'
)

# get track names to loop through
tracks = pivot_table_df.columns

# work_df is str->ms int
work = pivot_table_df.copy(deep=True)
# work.fillna(value=np.nan, inplace=True)

work.head()
```

```python
for track in tracks:
    # 2020 + 1
    for year in range(2000, 2021):
        current = pivot_table_df[track][year]
        if (current is not np.nan):
            work[track][year] = convert(track + str(year), pivot_table_df[track][year])

#work.head()
work
```

```python
#Helper for calculating percent difference as multiplier
def getPercentDiff(current, t1, t2):
    #t1 is previous, t2 current time all in Milliseconds
    try:
        if (t1 > t2):
            #t1 > t2; percent time decrease case
            return (1 - (t1- t2)/ t1)
        else:
            #t2 > t1; percent time increase case
            return (1 + (t2-t1) / t1)
    except:
        print(current, t1, t2)
```

```python
# change is the df to put calculations ; contains percentage difference

change = work.copy(deep=True)

#Helper for finding first valid index of each row
def first_valid_index(series):
    return series.first_valid_index()

first_valid_indices = change.apply(first_valid_index)
#print(first_valid_indices["Fuji Speedway"])
#print(first_valid_indices)
```

```python
#calculation for change df, gives final mulipliers for percent change

                
for track in tracks:
     # 2020 + 1
    for year in range(2000, 2021):
        track_i = pd.Index(tracks).get_loc(track) #index of track
        current = work[track][year]
        if (current is not np.nan):
            #check if it's the first entry in a new track
            if(first_valid_indices[track] == year):
                #no previous year, just started
                change[track][year] = 1 #1 == no difference calculted; default for first value
                previous = current
            else: 
                #should be year with previous year; eligible for calc
                #TODO: see if previous here 
                change[track][year] = getPercentDiff(track + str(year), previous, work[track][year])
                previous = current
    #print(work.set_index(track).diff()) #currently doing some random diff; not the correct calculations

# drop tracks that have multiple years skipped (aka multiple NaNs consecutively)
dropped = change.drop([
                       'Autodromo Internazionale del Mugello',
                       'Autódromo Internacional do Algarve',
                       'Bahrain International Circuit 1', 
                       'Baku City Circuit', 
                       'Buddh International Circuit',
                       'A1-Ring',
                       'Autodromo Enzo e Dino Ferrari',
                       'Autódromo Hermanos Rodríguez',
                       'Circuit Paul Ricard',
                       'Marina Bay Street Circuit',
                       'Circuit of the Americas',
                       'Fuji Speedway',
                       'Korean International Circuit',
                       'Valencia Street Circuit',
                       'Bahrain International Circuit',
                       'Sochi Autodrom',
                       'Istanbul Park',
                       'Circuit de Nevers Magny-Cours',
                       'Yas Marina Circuit',
                       'Shanghai International Circuit',
                       'Red Bull Ring',
                       'Nürburgring',
                       'Indianapolis Motor Speedway',
                       'Sepang International Circuit'
                      ], axis=1)
#dropped['avg'] = dropped.mean(axis=1)
dropped
```

```python
arr = dropped.to_numpy()
# print(arr)
data = []
# print(arr)
for idx, year in enumerate(arr):
    track_count = 0
    c = 0
    for avg in year:
        if avg is not np.nan:
            track_count += 1
            c+= avg
    val = 0
    if track_count > 0:
        val= c/track_count
    year = 2000 + idx
#     print(year, val)
    data.append([year, val])

print(data)
```
